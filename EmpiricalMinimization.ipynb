Expore the behaviour of the function:

\begin{equation}
R_T \leq  h + (T - h)\left(\sqrt{\frac{4}{h}\log(\frac{hk}{2})}+\frac{2}{h}\right)
\end{equation}

For different values of $T$ and $K$ I want to find the $h$ that minimizes the regret subject to the constraint that $T > h$ and everything is integer > 0

```python
def R(h,T,K):
    return h+(T-h)*(sqrt(4.0/h * log(h*K/2.0)) + 2.0/h)
```

```python
# Just to test that I've got minimize_scaler working correctly ...
from scipy.optimize import minimize_scalar
# minimize function with extra arguaments - ok that seems to work

def f(x,c=1):
    return (x-c)**2

res = minimize_scalar(f, bounds=(-10, 10), method='bounded',args=(0,))
print res.x
```

```python
T = 100
h = []
Kvals = arange(2,51,1)
for K in Kvals:
    res = minimize_scalar(R, bounds=(1, T-1), method='bounded',args =(T,K) )
    h.append(res.x)
plot(Kvals,h)
xlabel("K")
ylabel("Optimal h")
show()
```

```python
K = 50
h = []
Tvals = arange(50,1000,10)
for T in Tvals:
    res = minimize_scalar(R, bounds=(1, T-1), method='bounded',args =(T,K) )
    h.append(res.x)
plot(Tvals,h)
xlabel("T")
ylabel("Optimal h")
show()
    
```

```python
T = arange(100,1000,10)
K = arange(2,100,1)
T,K = meshgrid(T,K)
Z = zeros(T.shape)
for i in range(T.shape[0]):
    for j in range(T.shape[1]):
        z = minimize_scalar(R, bounds=(1, T[i,j]), method='bounded',args =(T[i,j],K[i,j]))
        Z[i,j] = z.x
```
