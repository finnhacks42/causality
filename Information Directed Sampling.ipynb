#Information Directed Sampling
This notebook contains an (attempted) implementation of Information Directed Sampling bandit algorithm as described in "Learning to Optimize via Information-Directed Sampling" for muli-armed Bernoulli bandits.


##Problem Setting
Assume we have actions $\{a_1...a_K\}$ where the mean rewards $X_i = P(y=1|a_i)$ for each arm $a_i$ are drawn i.i.d from a beta distribution $Beta(\alpha,\beta)$ We label the action with the highest expected reward $a^*$.

The goal is to choose actions so as to minimize the Bayesian Regret.

$I(X;Y) = D_{KL}(P(X,Y)||P(X)P(Y))$ is the mutual information (a measure of dependence) between variables $X$ and $Y$.

Information gain of action $a$ is defined as: $g(a) = I(A^*;Y(a)) = D_{KL}(P(A^* = a^*,Y=y|do(a)) || P(A^* = a)P(Y=y|do(a))$, 
 which (as a standard property of mutual information) equals $\mathbb{E}_{a^*}\left[D_{KL}(P(Y=y|do(a),A^*=a^*) || P(Y=y|do(a))\right]$

At each timestep, $t$, we select a policy, $\pi$, which assigns a probability to each action $\{a_1...a_K\}$ so as to minimize $\frac{\Delta_t(\pi)^2}{g(\pi)}$, where $\Delta_t(\pi)$ is the expected regret of policy $\pi$ and $g(\pi)$ is the information gain.






The first question is what does the random variable $A^*$ actually represent?

Now I want to calculate, for each arm, the probability that the true reward for that arm is equal to the optimal reward.

This equals the probability that the reward for this arm is $\geq$ to the reward for all the other arms which is given by:

\begin{equation}
\alpha_i = P\left(\cap_{j \neq i} \{p_j \leq p_i \} \right) = \int_0^1{f\left(p_i = x, p_j \leq x \; \forall j \neq i \right)}dx \\
=\int_0^1{f_i(x)P\left(\cap_{j \neq i} \{p_j \leq x | p_i = x \} \right)}dx = \int_0^1{f_i(x)\prod_{j \neq i}F_j(x)}dx
\end{equation}


```python
import scipy.integrate as integrate
n_samples = pow(2,10)+1
x = linspace(0,1,n_samples)
pdfs = asarray([p.pdf(x) for p in posteriors]) # creats a num_arms * n_samples array, where each row is the pdf for an arm
cdfs = asarray([p.cdf(x) for p in posteriors])
cdfs_dezerod = copy(cdfs)
cdfs_dezerod[cdfs==0] = 1

a = [integrate.simps(pdfs[i]*prod(cdfs,axis=0)/cdfs_dezerod[i],dx = 1/float(n_samples)) for i in range(num_arms)]
print(a)
```

## Calculate Information Gain
Now we can calculate the information gain for each arm

\begin{equation}
g(a_i) = \mathbb{E}_{j}\left[D_{KL}(P(Y|do(a_i),A^*=j) || P(Y|do(a_i))\right]
\end{equation}

The first term is the posterior over $y$ for arm $i$ given arm $j$ is the optimal arm. The second is just the posterior distribution over $y$ for arm $i$. 

\begin{equation}
P(Y|do(a_i)) = Bernoulli(\frac{success_a}{trials_a})\\
P(Y|do(a_i),A^*=j) = Bernoulli(M_{ji}) \; \text{ where $M_{ji} = E[p_i|p_k \leq p_j \; \forall{k}]$  ,the expected value of $p_i$ given $j$ is optimal} 
\end{equation}

If $i = j$:
\begin{equation}
M_{ii} = \int_0^1{x f(p_i = x|p_k \leq p_i \; \forall{k})dx} = \int_0^1{x \frac{P(p_k \leq x \; \forall{k}|p_i = x)f(p_i=x)}{P(p_k \leq p_i \; \forall{k})}dx} = \frac{1}{\alpha_i}\int_0^1{x \prod_{k \neq i}{F_k(x)}f_i(x)dx}
\end{equation}

If $i \neq j$:
\begin{equation}
M_{ij} = \int_0^1{x f(p_j = x|p_k \leq p_i \; \forall{k})dx} = \frac{1}{\alpha_i}\int_0^1{x f_j(x)P(p_k \leq p_i \; \forall{k}|p_j=x)dx} \\
\text{Now, } P(p_k \leq p_i \; \forall{k}|p_j=x) = \int_0^1{f(p_k \leq t \; \forall{k}, p_i = t|p_j=x)}dt = \begin{cases}
0 & \text{if $x > t$} \\
f_i(t)\prod_{k \neq i,j}F_k(t) & \text{otherwise} \end{cases}\\
\implies M_{ij} = \frac{1}{\alpha_i}\int_0^1f_i(t)\prod_{k \neq i,j}F_k(t)\left(\int_0^t{x f_j(x)dx}\right)dt
\end{equation}



```python
def M_ii(i):
    f = x*pdfs[i]*prod(cdfs,axis = 0)/cdfs_dezerod[i]
    #f2 = x*prod(cdfs[[j for j in range(num_arms) if j != i],:],axis=0)*pdfs[i]
    return 1.0/a[i]*integrate.simps(f,dx = 1/float(n_samples))
    
def M_ij(i,j):
    # calculate inner integral for each value of t
    f_inner = x*pdfs[j]
    F_inner_t = integrate.cumtrapz(f_inner,x,initial=0)
    
    # now do the outer integral and return the result ...
    f = pdfs[i]*prod(cdfs,axis=0)/cdfs_dezerod[i]*F_inner_t #TODO Check this line - should be dividing by cdfs_dezerod[j] too
    return 1.0/a[i]*integrate.simps(f,dx = 1/float(n_samples)) 
    
M = zeros((num_arms,num_arms))
for i in range(num_arms):
    for j in range(num_arms):
        if i == j:
            M[i,j] = M_ii(i)
        else:
            M[i,j] = M_ij(i,j)
```

The KL-divergence between Bernoulli distributions is given by:
\begin{equation}
D_{KL}(Bernoulli(p)||Bernoulli(q)) = p\log(\frac{p}{q})+(1-p)\log(\frac{1-p}{1-q}) = KL(p,q)
\end{equation}
So: 
\begin{equation}
\begin{aligned}
g(a_i) & = \mathbb{E}_{j}\left[D_{KL}(P(Y|do(a_i),A^*=j) || P(Y|do(a_i))\right]\\ 
& =  \mathbb{E}_{j}\left[D_{KL}(Bernoulli(M_{ji}) || Bernoulli(\frac{success_i}{trials_i})\right]\\ 
& =  \mathbb{E}_{j}\left[KL(M_{ji},\frac{\beta_{i1}}{\beta_{i1}+\beta_{i2}}) \right]\\
& = \sum_{j = 1}^K \alpha_j KL(M_{ji},\frac{\beta_{i1}}{\beta_{i1}+\beta_{i2}})
\end{aligned}
\end{equation}



```python
def kl(p,q):
    return p*log(p/q) + (1-p)*log((1-p)/(1-q))

g = [kl(M[:,i],(1+success[i])/float(trials[i]+2)).dot(a) for i in range(num_arms)]
print g
```

## Calculate Expected Regret for each arm
We calculate the expected regret for each arm

```python
est_p_best = sum([a[i]*M[i,i] for i in range(num_arms)])
delta = [est_p_best - (1+success[i])/float(trials[i]+2) for i in range(num_arms)]
print delta
```

## Select action to maximize information gain ratio

Finally, putting it all together we select our policy and sample an action from our policy

```python
from scipy.optimize import minimize_scalar

def best_policy(g,d):
    """ returns a pair of arms and the probability of playing the first one.
        equivelent to returning  a vector of probabilities for all the arms as we know that at most 2 of them are non-zero"""
    K = len(d)
    # we know the best policy has at most two non-zero components (see paper)
    best_ijq = (0,0,-1)
    min_info = None
        
    for i in xrange(K):
        for j in xrange(K):
            if i != j:
            # minimize (q*d[i]+(1-q)*d[j])^2/(q*g[i]+(1-q)*g[j]) subject to q in [0,1]
                objective = lambda q: pow(q*d[i]+(1-q)*d[j],2)/(q*g[i]+(1-q)*g[j])
                res = minimize_scalar(objective,bounds=(0,1),method='bounded')
                if min_info is None or res.fun < min_info:
                    min_info = res.fun
                    best_ijq = (i,j,res.x)
    assert best_ijq[2] >=0, "policy probability smaller than 0:"+str(best_ijq[2])           
    return best_ijq

p = best_policy(g,delta)
print p
```

```python
from scipy.stats import beta
from scipy.stats import bernoulli
from scipy.optimize import minimize_scalar

# f_i = pdf of X_i for each arm
# F_i = cdf of X_i for each arm

def kl(x,y):
    a = pow(x/y,x)
    b = pow((1-x)/(1-y),(1-x))
    result=log(a*b)
    if result < 0:
        result = 0
    #result = x*log(x/y)+(1-x)*log((1-x)/(1-y))
    assert result >= 0, "problem calculating KL({}||{}) output is: {}".format(x,y,result) 
    return result
     
class Arm:
    def __init__(self,a,b,n,indx):
        self.a = a
        self.b = b
        self.dx = 1.0/(n-1)
        self.grid = linspace(self.dx,1-self.dx,n-1)
        self._grid()
        self.indx = indx
        #print self.grid
        
    def update(self,success):
        if success:
            self.a +=1
        else:
            self.b +=1
        self._grid()
        
    
    def _grid(self):
        self.pdf = beta.pdf(self.grid,self.a,self.b) # the probability distribution function for the expected reward of this arm
        self.cdf = beta.cdf(self.grid,self.a,self.b)
        self.pdf = maximum(self.pdf,1e-323)
        self.cdf = maximum(self.cdf,1e-323)
        
        self.Q = self.dx*cumsum(self.grid*self.pdf)
           
    def Q(self):
        return self.Q
                
    def prob_optimal(self,F):
        result = self.dx*sum((self.pdf/self.cdf)*F)
        assert 0 < result < 1, "probability arm {} is optimal is: {}, not in [0,1] pdf:{},cdf:{}".format(self.indx,result,self.pdf,self.cdf)
        return result
    
    def expected_reward(self):
        result = self.a/float(self.a+self.b)
        assert 0 < result < 1, "expected reward of arm {} is:{}, not in [0,1]".format(self.indx,result)
        return result
    
    
    
    def expected_reward_given_i_best(self,F,arm_i):
        b = self.dx/arm_i.prob_optimal(F)
        
        if arm_i.indx == self.indx:
            result = b*sum((self.grid*self.pdf/self.cdf)*F)
        else:
            top = (arm_i.pdf*F)
            bottom = (self.cdf*arm_i.cdf)
            ratio = top/bottom
            # horrible hack to just get rid of nan
            if isnan(sum(ratio)):
                for i in xrange(len(ratio)):
                    if isnan(ratio[i]):
                        if i < len(ratio)-2:
                            ratio[i] = ratio[i+1]
                        else:
                            ratio[i] = ratio[i-1]
            result = b*sum(ratio*self.Q)
        if result >= 1 and result < 1.2:
            result = 0.999999999999
        assert 0 < result < 1, "expected reward of arm {} given arm {} is:{}, not in [0,1]".format(self.indx,arm_i.indx,result)#, b:{},ratio:{},Q:{},top:{},bottom:{}".format(self.indx,arm_i.indx,result,b,ratio,self.Q,top,bottom)
        return result
    
    def __str__(self):
        return "("+str(self.a)+","+str(self.b)+")"
    
    def times_played(self):
        return self.a+self.b
    
    

class IDSBandit:
    def __init__(self,a,b,grid_size,model):
        self.arms = [Arm(a,b,grid_size,i) for i in xrange(model.K)]
        self.F = [1]*(grid_size-1)
        self.updateF()
        self.model = model
        self.regret = 0
    
    def updateF(self):
        for arm in self.arms:
            self.F *= arm.cdf
            
    def information_gain(self):
        """ calculate the expected information gain from playing each arm in the next timestep """
        g = zeros((len(self.arms),))
        for k,arm in enumerate(self.arms):
            arm.info = 0
            E_k = arm.expected_reward()
            for i in xrange(len(self.arms)):
                prob_i_optimal = self.arms[i].prob_optimal(self.F)
                E_k_given_i_optimal = arm.expected_reward_given_i_best(self.F,self.arms[i])
                arm.info += prob_i_optimal*kl(E_k_given_i_optimal,E_k)
            g[k] = arm.info
        return g
    
    def expected_regret(self):
        """ caclulate the expected regret of playing each arm in the next timestep """
        best_expected_reward = max([arm.expected_reward() for arm in self.arms])
        delta = [best_expected_reward - arm.expected_reward() for arm in self.arms]
        return delta
        
    
    def best_policy(self,g,d):
        """ returns a pair of arms and the probability of playing the first one.
        equivelent to returning  a vector of probabilities for all the arms as we know that at most 2 of them are non-zero"""
        K = len(d)
        # we know the best policy has at most two non-zero components
        best_ijq = (0,0,-1)
        min_info = None
        
        for i in xrange(K):
            for j in xrange(K):
                # minimize (q*d[i]+(1-q)*d[j])^2/(q*g[i]+(1-q)*g[j]) subject to q in [0,1]
                objective = lambda q: pow(q*d[i]+(1-q)*d[j],2)/(q*g[i]+(1-q)*g[j])
                res = minimize_scalar(objective,bounds=(0,1),method='bounded')
                if min_info is None or res.fun < min_info:
                    min_info = res.fun
                    best_ijq = (i,j,res.x)
        assert best_ijq[2] >=0, "policy probability smaller than 0:"+str(best_ijq[2])           
        return best_ijq
                                                          
    def sample_action(self,policy):
        """ samples an arm according to the probabilities in the specified policy """
        ind = 1-bernoulli.rvs(policy[2])
        return policy[ind]
       
    
           
    def run(self,horizon):
        # track the evolution of a bunch of things over time
        K = len(self.arms)
        self.regret_t = zeros((horizon+1,)) # regret as a function of time
        self.ep_t = zeros((horizon+1,K)) # expected value of each arm
        self.played_t = zeros((horizon+1,K))
        self.d_t = zeros((horizon+1,K))
        self.g_t = zeros((horizon+1,K))
        
        
        
        for t in xrange(horizon):
            gain = self.information_gain() #I'm getting nan as a result for some of these...
            regret = self.expected_regret()
            policy = self.best_policy(gain,regret)
            arm = self.sample_action(policy)
            reward = self.model.play(arm)
            # update that arm
            self.arms[arm].update(reward)
           
            
            self.updateF()
            self.regret += model.best_payoff() - model.p[arm] # the true average payoff of the best arm vs the selected one
            self.regret_t[t+1] = self.regret
            #print "played arm",arm,"state:",self.arms[arm],"expected reward:",self.arms[arm].expected_reward()
            for k,arm in enumerate(self.arms):
                self.ep_t[t+1,k] = self.arms[k].expected_reward()
                self.played_t[t+1,k] = self.arms[k].times_played()
                self.d_t[t+1,k] = regret[k]
                self.g_t[t+1,k] = gain[k]
            
            
        return self.regret
    
            
class BetaRandomModel:
    """ the reward probaility for each arm is drawn from a beta prior"""
    def __init__(self,K,a,b):
        self.p = beta.rvs(a,b,size=K)
        self.K = K
        self.best = max(self.p)
              
    def play(self,arm):
        """ returns a reward sampled from the payoff distribution of the chosen arm"""
        return bernoulli.rvs(self.p[arm])
    
    def __str__(self):
        return str(self.p)
    
    def best_payoff(self):
        return self.best
    




          
```
