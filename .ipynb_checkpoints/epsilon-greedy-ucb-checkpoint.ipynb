#$\epsilon$-Greedy-Causal-UCB: Basic Idea

Setting is multiple independent binary variables $X_1...X_N$ cause binary reward variable $Y$. Available actions are to set any one variable to a given value - ie $K = 2N$. After we have played an action we observe the reward $Y$ and the values all the other variables took (odd setting I know)

1. for each arm keep two sets of confidence bounds. One based on observational data, the other based on interventions.
2. observe only (no action) with probability $\epsilon$ and update observational bounds on all variables selected in these rounds.
3. In remaining rounds, we select the arm with the highest upper confidence bound (considering the narrowest bound for each arm), and update interventional bounds for that variable. (Or I could thompson sample from the lowest variance posterior - possibly)
4. There may be some way of varying $\epsilon$ with time - huristically or based on something like information gain

```python
# imports etc
%pylab inline
%load_ext autoreload
%autoreload
from scipy.stats import bernoulli,beta,itemfreq
from numpy.random import randint
from bandit_utils import BernoulliEstimator,TrivialProbabilityModel,ucbBound
seed(42) # something for repeatability
```

## $\epsilon$-Greedy-Causal-UCB

```python
class EpsilonGreedyCausalUCBBandit:
    def __init__(self,model,alpha,epsilon,epsilonDecay):
        self.model = model
        self.obs = [BernoulliEstimator(1,1,alpha) for k in range(model.numArms)]
        self.intv = [BernoulliEstimator(1,1,alpha) for k in range(model.numArms)]
        self.epsilon = bernoulli(epsilon)
        
    def run(self,horizon):
        regret = zeros(horizon)
        for t in range(horizon):
            if self.epsilon.rvs()==1: # observe 
                X = self.model.sampleX() # randomly sample values of X
                Y = self.model.sampleYGivenX(X)  # sample Y from P(Y|X)
                
                for var,value in enumerate(X):   # get the corresponding action for each setting of X and update
                    action = self.model.toID(var,value)
                    self.obs[action].update(Y)
                    
                regret[t] = self.model.best - self.model.expectedY() # update regret
                        
            else: # do UCB
                upper = minimum([o.upper(t) for o in self.obs],[i.upper(t) for i in self.intv])
                action = argmax(upper)
                regret[t] = self.model.best - self.model.expectedYGivenAction(action)
                reward = self.model.sampleY(action)
                self.intv[action].update(reward)
        regret = cumsum(regret)
        self.t = horizon
        return regret
    
    def plot_bounds(self):
        f,ax = subplots(1,1)
        ub = minimum([o.upper(self.t) for o in self.obs],[i.upper(self.t) for i in self.intv])
        ax.plot(range(model.numArms),ub,'_',label="Upper Bound")
        ax.plot(range(model.numArms),[model.expectedYGivenAction(a) for a in range(model.numArms)],'o',color="red",label="True mean reward")
        ax.set_xlim(-1,10)
        ax.legend(loc="lower left",numpoints = 1)
        show()
    
```

## Explore Then Pick the best Bandit

```python
# Explore then pick the best

class ObserveExploitBandit:
    def __init__(self,model):
        self.model = model
        self.success = zeros(model.numArms)
    
    def run(self,horizon):
        regret = zeros(horizon)
        explore_h = self.h(horizon)
        for t in range(explore_h):
            X = self.model.sampleX() # returns a vector of length numVars specifying the value of each variable
            Y = self.model.sampleYGivenX(X)  # sample Y from P(Y|X)
                
            for var,value in enumerate(X):   # get the corresponding action for each setting of X and update
                action = self.model.toID(var,value)
                self.obs[action].update(Y)
                    
            regret[t] = self.model.best - self.model.expectedY() # update regret
        # for the remaining timesteps we use the arm with the highest estiated reward.
            
        
        
    def h(self,horizon):
        return int(min(ceil(horizon**(2/3.0)*(log(model.numArms*horizon))**(1/3.0)),horizon))
```

## UCB Bandit for comparison

```python
class UCBBandit: 
    def __init__(self,model,alpha):
        self.model = model
        self.estimators = [BernoulliEstimator(1,1,alpha) for k in range(model.numArms)]
             
    def run(self,horizon):
        regret = zeros(horizon)
        self.actions = zeros(horizon)
        for t in range(horizon):
            upper = [x.upper(t) for x in self.estimators]
            action = argmax(upper)
            self.actions[t] = action
            regret[t] = self.model.best - self.model.expectedYGivenAction(action)
            reward = self.model.sampleY(action)
            self.estimators[action].update(reward)
        regret = cumsum(regret)
        self.t = horizon
        return regret
    
    def plot_bounds(self):
        f,ax = subplots(1,1)
        ub = [x.upper(self.t) - x.mean() for x in self.estimators]
        ax.plot([model.expectedYGivenAction(a) for a in range(model.numArms)],'o',color="red",label="True mean reward")
        ax.errorbar(range(model.numArms), [x.mean() for x in self.estimators], yerr=ub, fmt='o',label="Estimated reward")
        ax.set_xlim(-1,10)
        ax.legend(loc="lower left",numpoints = 1)
        show()
```

## Compare Regret with time

```python
# consider having epsilon decay with time ...
num_vars = 10
horizon = 10000
trials = 20
alpha = 2.5
epsilon = 0.3

ucb = zeros((trials,horizon))
causal = zeros((trials,horizon))

for trial in range(trials):
    model = TrivialProbabilityModel(num_vars,.2)
    causalBandit = EpsilonGreedyCausalUCBBandit(model,alpha,epsilon)
    ucbBandit = UCBBandit(model,alpha)
    ucb[trial,:] = ucbBandit.run(horizon)
    causal[trial,:] = causalBandit.run(horizon)
    
f,ax = subplots(1,1,figsize=(15,10))
t = range(horizon)
ucb_se = std(ucb,axis=0)/sqrt(trials)
ucb_mean = mean(ucb,axis=0)
ax.plot(t,ucb_mean,label="UCB",color="green")
ucb_se = std(ucb,axis=0)/sqrt(trials)
ax.fill_between(t,ucb_mean-ucb_se,ucb_mean + ucb_se,alpha=0.3,color="green")

c_se = std(causal,axis=0)/sqrt(trials)
c_mean = mean(causal,axis=0)
ax.plot(t,c_mean,label="Causal", color = "blue")
ax.fill_between(t,c_mean-c_se,c_mean+c_se,alpha=0.3, color="blue")
ax.legend(loc="lower right")
ax.set_xlabel("Time period")
ax.set_ylabel("Cumulative Expected Regret")
show()

# Ok - this is roughly the behaviour I was expecting. 
# epsilon really does need to get smaller as we have explored more of the space ... 

```
