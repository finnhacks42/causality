##Estimators

Suppose we have a coin, with unknown bias $p$ which we wish to estimate. We flip the coin once and then with probability $q$ for $n$ timesteps.

$\mu_1 = \frac{\text{# of successes}}{\text{# of trials}}$

$\mu_2 = \frac{\text{# of successes}}{n q}$

```python
from scipy.stats import bernoulli
p = .5
q = .1
n = 100.0
trials = 1000

u1 = zeros((trials))
u2 = zeros((trials))
flips = zeros((trials))
for t in range(trials):
    flipped = bernoulli(q).rvs(n)
    flipped = sum(flipped)+1
    heads = bernoulli(p).rvs(flipped)
    u1[t] = sum(heads)/float(flipped)
    u2[t] = sum(heads)/float(q*n+1)
    flips[t] = flipped


f,ax = subplots(1,3,figsize=(12,5))
ax[0].hist(u1)
ax[0].set_xlabel("$\hat{\mu_1}$")
ax[0].set_title("standard estimator")
ax[1].hist(u2)
ax[1].set_xlabel("$\hat{\mu_2}$")
ax[1].set_title("importance estimator")
ax[2].hist(flips)
ax[2].set_title("number of flips")
show()

print "Mean flips",mean(flips), "Expected # flips",n*q+1
print "E[u1],",mean(u1), ", E[u2],",mean(u2)
print "V[u1],",var(u1), ", V[u2],",var(u2)
```

<h3> Markov inequality </h3>
for a non negative random variable $X$, 

$$P(X > t) \leq \frac{E[X]}{t}$$

```python
# example with the binomial distribution
from scipy.stats import binom
n, p = 100, 0.4
expectation = n*p
fig, ax = plt.subplots(1, 2)
x = np.arange(binom.ppf(0.5, n, p),binom.ppf(0.99999999999999, n, p)) # ppf function is inverse of CDF
ax[0].plot(x, binom.pmf(x, n, p), 'bo', ms=8, label='binom pmf')
#ax[0].vlines(x, 0, binom.pmf(x, n, p), colors='b', lw=5, alpha=0.5)

prob = binom.cdf(x,n,p)
ax[1].plot(x,prob,color="r")
ax[1].plot(x,1-expectation/x, color="g")
show()

# the bound is not especially tight for binomial distribution. Additional constraints eg, symetry would presumably make it tighter
```

```python
import math

ua1 = .5
ua0 = .1
pa = .1
u = pa*ua1+(1-pa)*ua0


# how many samples to take from each side
n = float(1000.0)
n1 = float(999.0)
n0 = float(n-n1)

# m1 ~ binomial(n1,pa)
# m0 ~ binomical(n0,(1-pa))

trials = 10000
uestimates = []
ua1est = []
ua0est = []
for t in range(trials):
    m1 = binom.rvs(n1, ua1)
    m0 = binom.rvs(n0,ua0)
    ua1hat = m1/n1
    ua0hat = m0/n0
    ua1est.append(ua1hat)
    ua0est.append(ua0hat)
    uhat = pa*ua1hat+(1-pa)*ua0hat
    uestimates.append(uhat)

fig, ax = plt.subplots(1, 3,figsize=(12,8))

ax[0].hist(ua1est)
ax[0].axvline(x=ua1, linewidth=2,color="red")

ax[1].hist(ua0est)
ax[1].axvline(x=ua0, linewidth=2,color="red")

ax[2].hist(uestimates)
ax[2].axvline(x=u, linewidth=2,color="red")
show()

# lets look at the distribution of the differences between the estimate and the true value
diff = [(x - u) for x in uestimates]
hist(diff)
show()


# calculate the bounds 
delta = .1
e = math.sqrt(.5*math.log(4/delta))*(pa*math.sqrt(1/n1)+(1-p)*math.sqrt(1/n0))
print e

# count the number of times |uhat -u| > e
beyond = sum(array(diff)>e)
print beyond/float(trials) # the fraction of trials where we go beyond the bound, should be < delta (as trials -> infinity)

# these bounds appear not too tight - we are always well below them


```
