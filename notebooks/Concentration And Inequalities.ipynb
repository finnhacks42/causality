```python
# Take a Bernoulli random variable and look at the distribution of its sum
import numpy as np
from numpy.random import binomial
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# variance of a bernoulli random variable
n = 100
p = 0.1
Sn = binomial(n,p,size=100000)

fig,ax = plt.subplots(1,2,figsize=(12,4))
ax[0].hist(Sn,normed=True,bins=25,alpha=0.5)
xlim = ax[0].get_xlim()
x = np.linspace(xlim[0],xlim[1],500)
ax[0].plot(x,norm(loc=n*p,scale=np.sqrt(n*p*(1-p))).pdf(x))
ax[0].set_xlabel("$S_n$")
ax[1].set_xlabel("$x$")
ax[0].set_ylabel("$P(S_n)$")
ax[1].set_ylabel("$P(S_n) > x$")

ax[1].hist(Sn,normed=True,bins=25,alpha=0.5,cumulative=-1)


print 1
# but variance is not enough. We need tail bounds. The probability that Sn > ?

# you could plot the mean with the standard deviation as some indication of the spread. 
# If the data were normal, we know the rough probabilities that the data is > some number of multiples of the standard deviation from the mean.
# (What is that function? - or its best approximation in closed form)

# we need to plot the cumulative...
```
