```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
n = 1000
u = np.random.normal(size=n)
z = u**2 + np.random.normal(scale=.1,size=n)
y = -u+np.random.normal(scale=0.1,size=n)
df = pd.DataFrame({"u":u,"z":z,"y":y})
```

```python
sns.pairplot(df)
```
